```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from statistics import median
```

```python
df_all = pd.read_csv('AmazonBS_other_info_2dep.csv')
```

```python
# 先把沒爬到的刪去
# 可先確認這些資料是否是需要的
for x in df_all.index:
    if df_all['all_rank'][x] == '_':
        df_all['all_rank'][x] = None

df_all = df_all.dropna(ignore_index=True)
```

```python
# 篩選format
# df_all['format']==  可換成想要看的foramt，但不要動到Paperback_df這個名稱
Paperback_df = df_all[df_all['format']=='Paperback']
```

```python
# 把str轉成int
for ind in Paperback_df.index:
  try:
    Paperback_df.all_rank[ind] = int(Paperback_df.all_rank[ind] \
                                                    .replace('#','') \
                                                    .replace(',',''))
  except:
    continue
```

```python
# 如果轉成int有例外狀況，檢視狀況
for ind in Paperback_df.index:
    if type(Paperback_df['all_rank'][ind]) != int:
        print(ind, Paperback_df['title'])
```

```python
# 以上狀況確定要刪除，就按下這個
for ind in Paperback_df.index:
    if type(Paperback_df['all_rank'][ind]) != int:
        Paperback_df['all_rank'][ind] = None

Paperback_df =  Paperback_df.dropna(subset = 'all_rank',ignore_index=True)
```

```python
# 找出第一層是什麼
dep_1 = []
for ind in Paperback_df.index:
  list_split  = Paperback_df.department[ind].split(' | ')
  dep_1.append(list_split[0])

Paperback_df['dep_1'] = dep_1
```

```python
# 第五名
def fif_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 5:
        return sorted_values[-5]
    else:
        return None
# 第十名
def ten_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 10:
        return sorted_values[-10]
    else:
        return None
#第20名 
def twen_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 20:
        return sorted_values[-20]
    else:
        return None
#第30名
def thirty_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 30:
        return sorted_values[-30]
    else:
        return None
#第40名   
def forty_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 40:
        return sorted_values[-40]
    else:
        return None
# 第40名
def fifty_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 50:
        return sorted_values[-50]
    else:
        return None
# 第100名
def handred_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 100:
        return sorted_values[-100]
    else:
        return None
# 第500名 
def five_handred_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 500:
        return sorted_values[-500]
    else:
        return None
# 第1000名
def thousand_larg(x):
    sorted_values = sorted(x, reverse=True)
    if len(sorted_values) >= 1000:
        return sorted_values[-1000]
    else:
        return None
```

```python
# 在第二層資料中，看第一層排名狀況並儲存
Paper_dep1 =  Paperback_df.groupby('dep_1')['all_rank'].agg([min,
                                                    fif_larg,
                                                    ten_larg,
                                                    twen_larg,
                                                    thirty_larg,
                                                    forty_larg,
                                                    fifty_larg,
                                                    handred_larg,
                                                    five_handred_larg,
                                                    thousand_larg,
                                                    ])
Paper_dep1.to_csv('Paper_dep1_agg.csv')
```

```python
# 檢視各類別第二層狀況並儲存
# 數字越小表示該類別越熱門
Paper_dep2 =  Paperback_df.groupby('department')['all_rank'].agg([min,
                                                    fif_larg,
                                                    ten_larg,
                                                    twen_larg,
                                                    thirty_larg,
                                                    forty_larg,
                                                    ])
Paper_dep2.to_csv('Paper_dep2_agg.csv')
```
