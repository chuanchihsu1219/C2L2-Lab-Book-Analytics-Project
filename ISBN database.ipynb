```python
import requests
import pandas as pd

# Define the publisher name
publisher_name = "HarperCollins"

# Set the page and pageSize parameters
page = 1
pageSize = 1000


# API endpoint URL with parameters
def construct_url(page: int):
    return f"https://api2.isbndb.com/publisher/{publisher_name}?page={page}&pageSize={pageSize}"


# Set the headers with the API key
headers = {"Authorization": API_KEY, "Accept": "application/json"}
responses = []
# Send GET request to the API
for page in range(1, 11):
    response = requests.get(construct_url(page), headers=headers)
    responses.append(response)
```

```python
print(response.content)
```

```python
import pandas as pd

dfs = []
# Assuming you have the JSON response stored in the 'response' variable
for response in responses:
    data = response.json()
    # Convert JSON to DataFrame
    df = pd.json_normalize(data["books"])
    # 選擇"title"和"isbn"列
    # df = df[["title", "isbn"]]
    dfs.append(df)
merged_df = pd.concat(dfs)
merged_df = merged_df.reset_index(drop=True)


# Display the DataFrame
print(merged_df.info())
```
